# What is RFM?
## RFM stands for:

### R = Recency: How recently did the customer make a purchase? (The number of days since the last purchase)
### F = Frequency: How often did the customer make a purchase? (Total number of purchases in a time period)
### M = Monetary: How much money did the customer spend? (Total revenue generated from the customer)

```python
reference_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
```

```python
reference_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,  
    'InvoiceDate': 'count',                                  
    'Total_Price': 'sum'
})

rfm.rename(columns={
    'InvoiceDate': 'Frequency',  
    '<lambda>': 'Recency'        
}, inplace=True)


rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,
    'InvoiceDate': 'count',   
})


recency = df.groupby('CustomerID')['InvoiceDate'].max().apply(lambda x: (reference_date - x).days)
frequency = df.groupby('CustomerID')['InvoiceDate'].count()
monetary = df.groupby('CustomerID')['Total_Price'].sum()

rfm = pd.DataFrame({
    'Recency': recency,
    'Frequency': frequency,
    'Monetary': monetary
})

print(rfm.head())
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'], 5, labels=[1, 2, 3, 4, 5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm['RFM_Score'] = (
    rfm['R_Score'].astype(str) +
    rfm['F_Score'].astype(str) +
    rfm['M_Score'].astype(str)
)
```

```python
def segment_customer(rfm_score):
    if rfm_score == '555':
        return 'Champion'
    elif rfm_score[0] == '5':
        return 'Loyal'
    elif rfm_score[2] == '5':
        return 'Big Spender'
    elif rfm_score == '111':
        return 'Lost'
    else:
        return 'Others'

rfm['Segment'] = rfm['RFM_Score'].apply(segment_customer)
```

```python
rfm_for_clustering = rfm[['Recency', 'Frequency', 'Monetary']].copy()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
rfm_scaled = sc.fit_transform(rfm_for_clustering)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km = KMeans(n_clusters=i, init='k-means++' ,n_init=10, random_state=42)
    km.fit_predict(rfm_scaled)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method - Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Inertia)')
```

```python
km = KMeans(n_clusters=4, init='k-means++',n_init=10 ,random_state=42)
rfm['Cluster'] = km.fit_predict(rfm_scaled)
```

```python
sns.scatterplot(x='Recency', y='Monetary', hue='Cluster', data=rfm, palette='viridis')
plt.title('Customer Segments (Recency vs Monetary)')
```
